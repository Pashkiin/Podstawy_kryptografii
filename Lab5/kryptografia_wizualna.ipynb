# Zaladowanie obrazka

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import random

# Load the image
secret_img = Image.open('sekret.png')

# Display the image
plt.imshow(secret_img)
plt.show()
```

```python
# Convert the image to black and white
bw_img = secret_img.convert('1')

# Display the image
plt.imshow(bw_img, cmap='gray')
plt.show()
print(bw_img.size)
```

```python
py_image_tBLW = np.array(bw_img)
```

```python
im1 = np.zeros(shape=(py_image_tBLW.shape[0], py_image_tBLW.shape[1]*2))
im2 = np.zeros(shape=(py_image_tBLW.shape[0], py_image_tBLW.shape[1]*2))

for y in range(py_image_tBLW.shape[0]):
    for x in range(py_image_tBLW.shape[1]):
        if py_image_tBLW[x][y] == 0:
            color = random.randint(0, 1)
            if color == 0:
                im1[x][y*2] = 1
                im1[x][y*2+1] = 0
                im2[x][y*2] = 1
                im2[x][y*2+1] = 0
            else :
                im1[x][y*2] = 0
                im1[x][y*2+1] = 1
                im2[x][y*2] = 0
                im2[x][y*2+1] = 1
        else:
            color = random.randint(0, 1)
            if color == 0:
                im1[x][y*2] = 0
                im1[x][y*2+1] = 1
                im2[x][y*2] = 1
                im2[x][y*2+1] = 0
            else:
                im1[x][y*2] = 1
                im1[x][y*2+1] = 0
                im2[x][y*2] = 0
                im2[x][y*2+1] = 1
    plt.subplot(1,2,1)
    plt.imshow(im1, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(im2, cmap='gray')
    

            
```

```python
im_result = np.zeros(shape=(py_image_tBLW.shape[0], py_image_tBLW.shape[1]*2))

for y in range(py_image_tBLW.shape[0]):
    for x in range(py_image_tBLW.shape[1]):
        im1p1 = im1[x][y*2]
        im1p2 = im1[x][y*2+1]
        im2p1 = im2[x][y*2]
        im2p2 = im2[x][y*2+1]
        if(im1p1 == 1 and im1p2 == 0 and im2p1 == 1 and im2p2 == 0):
            im_result[x][y*2] = 1
            im_result[x][y*2+1] = 0
        elif(im1p1 == 0 and im1p2 == 1 and im2p1 == 0 and im2p2 == 1):
            im_result[x][y*2] = 0
            im_result[x][y*2+1] = 1
        else:
            im_result[x][y*2] = 1
            im_result[x][y*2+1] = 1

plt.imshow(im_result, cmap='gray')
```

```python
im_result = np.zeros(shape=(py_image_tBLW.shape[0], py_image_tBLW.shape[1]))

for y in range(py_image_tBLW.shape[0]):
    for x in range(py_image_tBLW.shape[1]):
        im1p1 = im1[x][y*2]
        im1p2 = im1[x][y*2+1]
        im2p1 = im2[x][y*2]
        im2p2 = im2[x][y*2+1]
        if(im1p1 == 1 and im1p2 == 0 and im2p1 == 1 and im2p2 == 0) or (im1p1 == 0 and im1p2 == 1 and im2p1 == 0 and im2p2 == 1):
            im_result[x][y] = 0
        else:
            im_result[x][y] = 1

plt.imshow(im_result, cmap='gray')
```
